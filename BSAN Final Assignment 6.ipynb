# COLUMNS DESCRIPTION

CAR ID:
A unique identifier assigned to each car in the dataset. This ensures that each entry is distinct and can be referenced individually.

PARENT COMPANY:
The name of the parent company that owns the car brand. This helps in analyzing the distribution of vehicles across different automotive conglomerates.

MAKE ORIGIN:
The country or region where the car's brand originates. This is useful for geographical analysis and understanding global market presence.

BRAND:
The specific brand or manufacturer of the car. This column is essential for brand-specific analysis, such as brand popularity and market share.

MODEL/CLASS:
The model or class designation of the car, often indicating the specific type or series within a brand's lineup. This helps differentiate between various models produced by the same brand.

CAR NAME:
The full name of the car, often combining the brand, model, and sometimes additional identifiers like trim levels. This column provides a detailed reference for each vehicle.

MODEL (YEAR):
The year the car model was released. This is important for understanding the age of the car and for time-series analysis of model releases.

CAR TYPE:
The classification of the car, such as sedan, SUV, truck, etc. This column helps in analyzing the distribution of different car types and their market trends.

IMAGE:
A link or path to an image of the car provided by dealer.

PRICE RANGE:
The general price category of the car (e.g., low, luxury). This provides a broad overview of the market segment in which the car falls.

PRICE($):
The exact price of the car in US dollars. This column is crucial for price analysis, comparisons, and building predictive models for pricing.

AGE OF CAR:
The age of the car, calculated from the model year to the present. This helps in analyzing depreciation, resale value, and market demand for different age groups.

STOCK TYPE:
Indicates whether the car is new, used, or certified. This information is essential for analyzing market segments and understanding the condition of vehicles in the dataset.

MILEAGE:
The total distance the car has traveled, typically measured in miles. This is a key factor in determining the car’s condition and market value.

RATING:
The average user rating for the car, usually on a scale (e.g., 1 to 5). This provides insight into customer satisfaction and perceived quality.

REVIEW:
The number of reviews the car has received.

DEALER NAME:
The name of the dealership selling the car. This is useful for analyzing dealer performance and market presence.

DEALER LOCATION (CITY):
The city where the dealership is located. This allows for geographical analysis at the city level.

DEALER LOCATION (STATE):
The state where the dealership is located. This allows for geographical analysis at the state level and can help identify regional trends.

```python
df = pd.read_csv("CarsData.csv")

print(df.dtypes)
```

```python
#Changing some columns to be catogorical

cols_to_category = ["MAKE ORIGIN", "BRAND", "CAR TYPE", "PRICE RANGE",
    "STOCK TYPE", "DEALER LOCATION (CITY)", "DEALER LOCATION (STATE)", "DEALER NAME"]

for c in cols_to_category: 
    df[c] = df[c].astype("category")

print(df.dtypes)
```

```python
# Changing/updating data types for necessary columns

if "MILEAGE" in df.columns:
    df["MILEAGE"] = df["MILEAGE"].fillna(df["MILEAGE"].mean())
if "RATING" in df.columns:
    df["RATING"] = df["RATING"].fillna(df["RATING"].mean())
if "REVIEW" in df.columns:
    df["REVIEW"] = df["REVIEW"].fillna(df["REVIEW"].mean())

cat_cols = ["PRICE RANGE", "STOCK TYPE", "MAKE ORIGIN", "BRAND",
    "CAR TYPE", "DEALER LOCATION (CITY)", "DEALER LOCATION (STATE)"]

for c in cat_cols:
    if c in df.columns:
        df[c] = df[c].astype("object")   
        df[c] = df[c].fillna("Unknown")  
        df[c] = df[c].astype("category") 

if "DEALER NAME" in df.columns:
    df["DEALER NAME"] = df["DEALER NAME"].astype("object")
    df["DEALER NAME"] = df["DEALER NAME"].fillna("No Dealer Info")

print(df.isna().sum())
```

```python
import pandas as pd
import statsmodels.api as sm

# Drop rows with missing important values
df_model = df[['PRICE(K$)', 'MILEAGE', 'MODEL_YEAR', 'AGE OF CAR', 'RATING',
               'BRAND', 'MAKE ORIGIN', 'STOCK TYPE', 'CAR TYPE', 'REGION']].dropna()
```

```python
# Multiple Linear (Numeric) regression

X1 = df_model[['MILEAGE', 'MODEL_YEAR', 'AGE OF CAR', 'RATING']]
y = df_model['PRICE(K$)']

X1 = sm.add_constant(X1)
model1 = sm.OLS(y, X1).fit()

print(model1.summary())
```

Model 1 Results:

In Model 1, I looked only at the numeric features of the cars, like mileage, model year, age, and rating. The model didn’t explain a lot of the price variation on its own, with an adjusted R² of only 0.131, meaning it captured about 13% of what affects resale prices. Still, all four variables were statistically significant. Mileage lowered the price as expected, and newer model years increased it. Cars with higher ratings also tended to sell for more. Age of car showed a positive effect here, which likely reflects collectible or vintage vehicles in the dataset. Overall, Model 1 shows that while basic characteristics matter, they clearly don’t tell the whole story.

```python
# Multiple Linear Regression (Numeric & categorical with dummies)

import statsmodels.api as sm
from statsmodels.stats.anova import anova_lm

df_model = df[['PRICE(K$)', 'MILEAGE', 'MODEL_YEAR', 'AGE OF CAR', 'RATING',
               'BRAND', 'MAKE ORIGIN', 'STOCK TYPE', 'CAR TYPE', 'REGION']].dropna()

X_num = df_model[['MILEAGE', 'MODEL_YEAR', 'AGE OF CAR', 'RATING']]

# Dummies
BR_dummies = pd.get_dummies(df_model["BRAND"],        prefix="BR", dtype=float)
MO_dummies = pd.get_dummies(df_model["MAKE ORIGIN"],  prefix="MO", dtype=float)
ST_dummies = pd.get_dummies(df_model["STOCK TYPE"],   prefix="ST", dtype=float)
CT_dummies = pd.get_dummies(df_model["CAR TYPE"],     prefix="CT", dtype=float)
RG_dummies = pd.get_dummies(df_model["REGION"],       prefix="RG", dtype=float)

# Combine everything
X2 = pd.concat([X_num, BR_dummies, MO_dummies, ST_dummies, CT_dummies, RG_dummies], axis=1)

y = df_model["PRICE(K$)"].astype(float)

X2 = sm.add_constant(X2)
model2 = sm.OLS(y, X2).fit()
print(model2.summary())
```

Model 2 results:

Model 2 added the categorical factors like brand, car type, stock type (new/used/certified), region, and make origin. Once these were included, the model improved a lot, with an adjusted R² jumping to 0.633. This means Model 2 explains more than half of the variation in resale prices, showing how important these categories are. Many brands had huge impacts on price, with luxury brands like Ferrari, Lamborghini, and Bugatti raising resale value significantly, while more affordable brands lowered it. Certified vehicles tended to sell for more than standard used ones, and certain car types like hybrids and electric vehicles also influenced price. Mileage and model year stayed significant as expected, but now the model paints a much fuller picture. Overall, Model 2 shows that brand, category, and market factors play a huge role in how much a car sells for.

```python
# Anova Results 

from statsmodels.stats.anova import anova_lm

anova_results = anova_lm(model1, model2)
print(anova_results)
```

Anova Results:

The ANOVA test compared Model 1 and Model 2 to see whether adding the categorical variables actually made the model better. The results showed a dramatic improvement, with an extremely high F-value and a p-value of essentially 0.0. This means the difference between the two models is statistically significant. In other words, Model 2 does a much better job predicting resale prices than Model 1. 

From the research I did, my understanding is that the NaNs in the first row are normal......they just mean Model 1 is the baseline and has nothing to compare to yet. What really matters is the second row, which shows that Model 2 explains significantly more variation. This confirms that including brand, car type, and similar features makes the model far more accurate.
